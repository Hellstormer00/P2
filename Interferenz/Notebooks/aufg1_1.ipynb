```python
import os
DIR = os.path.abspath('')
DATA = os.path.join(DIR, '../Data')
PLOTS = os.path.join(DIR, '../Plots')

sns.set_style('whitegrid')

data_blau = pd.read_csv(os.path.join(DATA, 'aufg1_1_blau.csv'), comment='#', engine='python')
data_gelb = pd.read_csv(os.path.join(DATA, 'aufg1_1_gelb.csv'), comment='#', engine='python')
data_gelb
```

```python
def prep(data):
    data['r_k[mm]^2'] = (abs(data['r_k_1[mm]'] - data['r_k_2[mm]'])/2)**2
    data['k'] = data.index + 1
    
def plot(data, name):
    sns.regplot('k', 'r_k[mm]^2', data)
    plt.ylabel('$r_k^2[mm^2]$')
    plt.savefig(os.path.join(PLOTS, name), dpi=200)
    
def R(data, l0):
    p = np.polyfit(data['k'], data['r_k[mm]^2'], 1)[0]  # = l0 * R
    print("R =", p/l0, "mm")
    return p/l0
```

```python
prep(data_blau)
plot(data_blau, 'aufg1_1_blau.png')
R(data_blau, 465e-6)
```

```python
prep(data_gelb)
plot(data_gelb, 'aufg1_1_gelb.png')
R(data_gelb, 590e-6)
```

```python
print("R =", (R(data_gelb, 590e-6) + R(data_blau, 465e-6))/2, "mm")
```
