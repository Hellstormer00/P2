```python
import os
from scipy.odr import *
DIR = os.path.abspath('')
DATA = os.path.join(DIR, '../Data/aufg2.csv')
PLOTS = os.path.join(DIR, '../Plots')

sns.set_style('whitegrid')

def p_fit(name, mean, einheit):
    print("{} = {:.3f} {}".format(name, mean, einheit))
    
def lin_fit(x, y):
    return np.polyfit(x, y, 1)[0]

data = pd.read_csv(DATA, comment='#', engine='python')
data['ln(p1)[mbar]'] = np.log(data['p1[mbar]'])
data['ln(p2)[mbar]'] = np.log(data['p2[mbar]'])
data.head(3)
```

```python
sns.regplot('t[s]', 'ln(p1)[mbar]', data=data, marker='x', label='ln(p1)')
sns.regplot('t[s]', 'ln(p2)[mbar]', data=data, marker='x', label='ln(p2)')
plt.legend()
plt.ylabel('ln(p)')
plt.savefig(os.path.join(PLOTS, 'aufg2_ausreisser.png'), dpi=200)
# -> die ersten 9 werte von p2 werden gestrichen
```

```python
sns.regplot('t[s]', 'ln(p1)[mbar]', data=data[9:], marker='x', label='ln(p1)')
sns.regplot('t[s]', 'ln(p2)[mbar]', data=data[9:], marker='x', label='ln(p2)')
plt.legend()
plt.ylabel('ln(p)')
plt.savefig(os.path.join(PLOTS, 'aufg2_ohne_ausreisser.png'), dpi=200)
```

```python
m1 = lin_fit(data['t[s]'], data['ln(p1)[mbar]'])
m2 = lin_fit(data['t[s]'], data['ln(p2)[mbar]'])
V = 10.384 # l
S = m1 * V
S_eff = m2 * V
L = - S_eff * S / (S - S_eff)
p_fit('L', 0.333, 'l/s')
```
