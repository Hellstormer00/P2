```python
import os
from scipy.odr import *
DIR = os.path.abspath('')
DATA = os.path.join(DIR, '../Data/aufg4.csv')
PLOTS = os.path.join(DIR, '../Plots')

sns.set_style('whitegrid')

def p_fit(name, mean, einheit):
    print("{} = {:.3e} {}".format(name, mean, einheit))
    
def lin_fit(x, y):
    return np.polyfit(x, y, 1)[0]

data = pd.read_csv(DATA, comment='#', engine='python')
data['ln(p)'] = np.log(data['p[mbar]'])
data.head(3)
```

```python
sns.scatterplot('t[s]', 'ln(p)', data=data, marker='x')
plt.savefig(os.path.join(PLOTS, 'aufg4_alldata.png'))
```

```python
sns.regplot('t[s]', 'ln(p)', data=data[6:], marker='x')
plt.savefig(os.path.join(PLOTS, 'aufg4_regression.png'))
```
