```python
import os 
from scipy.signal import argrelmin, argrelmax
DIR = os.path.abspath('')
DATA = os.path.join(DIR, '../Data')
PLOTS = os.path.join(DIR, '../Plots')
U = 2  # V

def read(T):
     data = pd.read_csv(os.path.join(DATA, str(T)+".csv"), comment='#')[::100].astype(float)
     data['']
    
def plot(T, start):
    data = read(T)[start:]
    maxes = argrelmax(np.array(data['U_A']), order=2)[0]
    plt.plot(data['U_B'], data['U_A'])
    plt.scatter(data.iloc[maxes]['U_B'], data.iloc[maxes]['U_A'], label='Maxima')
    plt.xlabel('$U_B$[V]')
    plt.ylabel('$U_A$[V]')
    plt.savefig(os.path.join(PLOTS, str(T)+".png"), dpi=200)
    
def dU(T, start):
    data = read(T)[start:]
    maxes = argrelmax(np.array(data['U_A']), order=2)[0]
    return np.mean(np.diff(maxes))
```

```python
plot(150, 10)
dU(150, 10)
```
